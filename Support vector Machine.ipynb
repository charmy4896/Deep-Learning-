```
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

from matplotlib import pyplot as plt
import pandas as pd
from copy import deepcopy
import numpy as np

def GoogleDriveAuth(ID):
    auth.authenticate_user()
    gauth = GoogleAuth()
    gauth.credentials = GoogleCredentials.get_application_default()
    drive = GoogleDrive(gauth)
    file = drive.CreateFile({'id': ID})
    return file
```

```
File_ID = '1gxAP1e97ynITr4p2m3XOVkAwFwRGyYo5'
file_name = GoogleDriveAuth(File_ID)
file_name.GetContentFile('ex2data1-logistic.xls')
df = pd.read_excel("ex2data1-logistic.xls") #load the dataset

# importing required libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
# reading csv file and extracting class column to y. 
a = np.array(df) 
y  = a[:,-1] # classes having 0 and 1 
x=a[:,:-1]
```

```
# import support vector classifier 
from sklearn.svm import SVC # "Support Vector Classifier" 
clf = SVC(kernel='linear') 
  
# fitting x samples and y classes 
clf.fit(x, y) 

#print(clf.predict([[34.62365962 ,78.02469282]]) )  
#print(clf.predict([[60.18259939, 86.3085521 ]]) )

# plotting scatter 
plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='spring') 
plt.xlim(x.min(), x.max())

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.title('LINEAR')
plt.show()


```
