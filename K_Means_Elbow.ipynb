```
import pandas as pd 
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

def GoogleDriveAuth(ID):
    auth.authenticate_user()
    gauth = GoogleAuth()
    gauth.credentials = GoogleCredentials.get_application_default()
    drive = GoogleDrive(gauth)
    file = drive.CreateFile({'id': ID})
    return file
```

```
File_ID = '0B1uNIZBUi1VzSURMZ0RISEdhNS1oX1VqRzFDSlFnS2FCQ1dr'
file_name = GoogleDriveAuth(File_ID)
file_name.GetContentFile('kmeans_data.xlsx')
df = pd.read_excel("kmeans_data.xlsx") #load the dataset
data = df.values[:, 0:2]

error = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(data)
    kmeanModel.fit(data)
    error.append(sum(np.min(cdist(data, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0])

plt.plot(K, error)
plt.xlabel('K=Number Of Clusters')
plt.ylabel('Cost')
plt.title('The Elbow Method to find optimal K on K_MEANS Dataset')
plt.show()

cost = []
c=[]
for i in range(100):
  kmeanModel = KMeans(n_clusters=3).fit(data)
  cost.append(sum(np.min(cdist(data, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0])
  c.append(kmeanModel.cluster_centers_)
ii=cost.index(min(cost))
print('Minimum Cost is at iteration',ii)
print(c[ii])

for i in range(data.shape[0]):
    plt.scatter(data[i, 0], data[i,1], s=7)
plt.scatter(c[ii][:,0], c[ii][:,1], marker='*', c='g', s=150)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Clusters')
plt.show()
```
