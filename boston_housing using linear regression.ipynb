```
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

def GoogleDriveAuth(ID):
    auth.authenticate_user()
    gauth = GoogleAuth()
    gauth.credentials = GoogleCredentials.get_application_default()
    drive = GoogleDrive(gauth)
    file = drive.CreateFile({'id': ID})
    return file


File_ID = '1Bzo2OedJw_eCf3j47RrOUFEraX5kJUPe'
file_name = GoogleDriveAuth(File_ID)
file_name.GetContentFile('boston_housing.csv')

import pandas as pd
import numpy as np
import math

data = pd.read_csv('boston_housing.csv')
df = pd.DataFrame(data)

mean=df.mean(axis=0)
std=df.std(axis=0)
df=df-mean
df=df.div(std)

Y_mean=(mean['MEDV'])
Y_Std=(std['MEDV'])

Y= df['MEDV']

X=df[['CRIM', 'ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
X.insert(0, 'X0', 1)

features=len(X.columns)
theta=[0 for i in range(features)]
alpha=0.025
m=X.X0.count()

COST_VALUE=[]
for i in range (1000):
  h_theta=np.matmul(X,np.transpose(theta))
  sub=h_theta-Y
  cost_value=(np.matmul(np.transpose(sub),sub))/(2*m)
  COST_VALUE.append(cost_value)
  theta=theta-(np.matmul(np.transpose(sub),X))*(alpha/m)

X_test1=[1,0.085, 13.0, 10.5, 1.0, 0.8, 4.78, 39.0, 5.5, 5.5, 331.0, 13.3,390.5, 17.71]
X_test2 = [1,0.17899, 0, 9.69, 0, 0.585, 5.67, 28.8, 2.7986, 6, 391, 19.2, 393.29, 17.6]
X_test1 = (X_test1 -np.mean(X_test1))/np.std(X_test1) 
X_test2 = (X_test2-np.mean(X_test2))/np.std(X_test2) 

ans=np.matmul(X_test1,np.transpose(theta))
print('X1 predicted values=',end=' ')
print((ans*Y_Std)+Y_mean)


ans=np.matmul(X_test2,np.transpose(theta))
print('X2 predicted values=',end=' ')
print((ans*Y_Std)+Y_mean)

print('Weight Vector')
print(theta)
print('COST FUNCTION')
print(COST_VALUE)
print()
```
